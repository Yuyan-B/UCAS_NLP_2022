```python
import pandas as pd
from sklearn.model_selection import KFold, StratifiedKFold
```

```python
raw_data = pd.read_csv('./data/starting_ki/train_all_tasks.csv')
raw_data
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
for i, (train_index, val_index) in enumerate(kf.split(raw_data, raw_data['label_sexist'])):
    train_set = raw_data.iloc[train_index]
    val_set = raw_data.iloc[val_index]
    # print(train_set[train_set['label_sexist'] == 'sexist'].shape)
    # print(train_set[train_set['label_sexist'] == 'not sexist'].shape)
    train_set.to_csv(f'./data/5fold/train{i}.csv', index=False)
    val_set.to_csv(f'./data/5fold/val{i}.csv', index=False)
```
